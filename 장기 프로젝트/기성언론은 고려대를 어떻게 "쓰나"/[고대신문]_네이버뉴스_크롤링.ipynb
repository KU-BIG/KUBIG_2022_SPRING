```python
# Packages for data crawling
from selenium import webdriver
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.keys import Keys
import time
import sys
import requests

import re
import pandas as pd
from bs4.builder import TreeBuilder
from bs4 import BeautifulSoup
import urllib.request
from urllib import parse
from tqdm import *
from selenium.webdriver.common.by import By

!apt-get update
!apt install chromium-chromedriver
print('All packages Ready')
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument("--single-process")
options.add_argument("--window-size=1920,1080")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("disable-gpu")

options.add_argument("user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36")
driver = webdriver.Chrome('chromedriver', options=options)

delay_time = 1
driver.implicitly_wait(delay_time)
```

```python
from pandas.core.base import NoNewAttributesMixin
total_page = 100

# 저장한 df, list 선언
news_df = pd.DataFrame(columns = ['press', 'title',
                                  'content', 'date', 'url'])
df_title=[]
df_press=[]
df_content=[]
df_date=[]
df_url=[]

# 검색 키워드 입력
query = input('검색 키워드 : ') 
query = query.replace(' ', '+') 

# 각 페이지에 대해 반복문 실행
for i in range(1, 22, 10):
  for j in range(1, 11):
    url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%s&start=%d' % (query, i)
    driver.get(url)
    # 기사 제목
    try:
      title = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div[2]/div/div[1]/section/div/div[2]/ul/li[%d]/div/div/a' % j)
      title = title.text
      df_title.append(title)
    except:
      title = None
    # 언론사
    try:
      press = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div[2]/div/div[1]/section/div/div[2]/ul/li[%d]/div/div/div[1]/div[2]/a[1]' % j)
      press = press.text
      press = press.replace('언론사 선정', '')
      df_press.append(press)
    except:
      press = None
    # 작성 일시
    try:
      date = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div[2]/div/div[1]/section/div/div[2]/ul/li[%d]/div[1]/div/div[1]/div[2]/span' % j)
      date = date.text
      df_date.append(date)
    except:
      date = None
    # 기사 URL
    try:
      url = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div[2]/div/div[1]/section/div/div[2]/ul/li[%d]/div[1]/div/a' % j).get_attribute('href')
      df_url.append(url)
    except:
      url = None
    # 기사 본문
      

for n in range(len(df_title)):
  append_data = {'press' : df_press[n], 'title' : df_title[n],
                 'date' : df_date[n], 'url' : df_url[n]}
  news_df = news_df.append(append_data, ignore_index=True)
news_df
```

```python
# 크롤링 내용 저장
news_df.to_csv('navernews_crawling.csv')
```
