## 3. 리뷰별 품사로 태깅된 표현들 저장하기

```python
mag_va = []
nng_va = []
mag_vcp = []
nng_vcp = []
mag_vv = []
nng_vv = []
nng_jk_va = []
nng_jx_va = []
nng_nng_vcp = []
nng_jk_vv = []
nng_jx_vv = []


for i in range(len(review)):
  try:
    string = review[i]
    pos = kkma.pos(string)

    ##
    for k in range(len(pos)-1):
      if (pos[k][1] == 'MAG') and (pos[k+1][1] == 'VA'):
        word = pos[k][0]+' '+pos[k+1][0]
        mag_va.append(word)
        
      if (pos[k][1] == 'NNG') and (pos[k+1][1] == 'VA'):
        word = pos[k][0]+' '+pos[k+1][0]
        nng_va.append(word)
   
      if (pos[k][1] == 'MAG') and (pos[k+1][1] == 'VCP'):
        word = pos[k][0]+' '+pos[k+1][0]
        mag_vcp.append(word)
  
      if (pos[k][1] == 'NNG') and (pos[k+1][1] == 'VCP'):
        word = pos[k][0]+' '+pos[k+1][0]
        nng_vcp.append(word)

      if (pos[k][1] == 'MAG') and (pos[k+1][1] == 'VV'):
        word = pos[k][0]+' '+pos[k+1][0]
        mag_vv.append(word)
    
    ###
    for k in range(len(pos)-2):
      if (pos[k][1] == 'NNG') and (pos[k+1][1] == 'JK') and (pos[k+2][1] == 'VA') :
        word = pos[k][0]+pos[k+1][0] + ' ' + pos[k+2][0]
        nng_jk_va.append(word)

      if (pos[k][1] == 'NNG') and (pos[k+1][1] == 'JX') and (pos[k+2][1] == 'VA') :
        word = pos[k][0]+pos[k+1][0] + ' ' + pos[k+2][0]
        nng_jx_va.append(word)

      if (pos[k][1] == 'NNG') and (pos[k+1][1] == 'NNG') and (pos[k+2][1] == 'VCP') :
        word = pos[k][0]+' '+pos[k+1][0] + ' ' + pos[k+2][0]
        nng_nng_vcp.append(word)

      if (pos[k][1] == 'NNG') and (pos[k+1][1] == 'JK') and (pos[k+2][1] == 'VV') :
        word = pos[k][0]+pos[k+1][0] + ' ' + pos[k+2][0]
        nng_jk_vv.append(word)

      if (pos[k][1] == 'NNG') and (pos[k+1][1] == 'JX') and (pos[k+2][1] == 'VV') :
        word = pos[k][0]+pos[k+1][0] + ' ' + pos[k+2][0]
        nng_jx_vv.append(word)




  except :
    pass

```

```python
word_list = pd.Series(mag_va)
count = word_list.value_counts()

exp = []
cnt = []
for i in range(len(count)):
  exp.append(count.index[i])
  cnt.append(count[i])

data = {
    'expression':exp,
    'count':cnt
}
df_mag_va = pd.DataFrame(data)

df_mag_va.to_csv('/content/drive/MyDrive/식당추천시스템/MAG_VA_count.csv',index=False, encoding='utf-8-sig')
```
